```python
#from sklearn import linear_model
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
experiencia = pd.Series ([1.1,1.3,1.5,2.0,2.2,2.9,3.0,3.2,3.2,3.7,3.9,4.0,4.0,4.1,4.5,4.9,5.1,5.3,5.9,6.0,6.8,7.1,7.9,8.2,8.7,9.0,9.5,9.6,10.3,10.5])
experiencia
salario = pd.Series([39343.00,46205.00,37731.00,43525.00,39891.00,56642.00,60150.00,54445.00,64445.00,57189.00,63218.00,55794.00,55957.00,57081.00,61111.00,67938.00,66029.00,83088.00,81363.00,93940.00,91738.00,98273.00,101302.00,113812.00,109431.00,105582.00,116969.00,112635.00,122391.00,121872.00])
salario
df= pd.DataFrame({"Experiencia": experiencia, "Salario":salario})
df
```

```python
# division de los datos (en de entrenamiento del 75 al 80%) y (de prueba del 20 al 25%) 

X_experiencia=experiencia
y_salario= salario
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X_experiencia, y_salario,test_size= 0.2, random_state = 42)
print("Núm de datos de entrenamiento Experiencia:  ", X_train.shape)#datos de entranamiento experiencia
print("Núm de datos de prueba Experiencia:  ",X_test.shape)# datos de testeo experiencia
print("Núm de datos de entrenamiento Salario:  ",y_train.shape)  #datos de entranamiento salario
print("Núm de datos de prueba Salario:  ",y_test.shape) # datos de testeo salario
```

#**ML SUPERVISADOS**
Los modelos de Machine Learning supervisados son aquellos que se entrenan utilizando un conjunto de datos que ya están etiquetados o categorizados. Es decir, se les proporciona al modelo tanto las características (features) de los datos como la respuesta o etiqueta correspondiente a cada dato.

El objetivo de estos modelos es aprender una función que pueda predecir la respuesta o etiqueta correcta para datos nuevos y desconocidos. Por lo tanto, estos modelos se utilizan para problemas de predicción y clasificación.

```python
#Creación del modelo de regresión lineal
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X_train =X_train.values.reshape(-1,1)  # una vez que lo hemos corrido ya no es necesario usarlo de nuevo por eso se deja en comentario

lin_model = LinearRegression()
lin_model.fit(X_train, y_train)  #apliclamos el metodo de regresión linial a los datos de entrenamiento
```

```python
from sklearn.metrics import r2_score
X_test =X_test.values.reshape(-1,1)  # una vez que lo hemos corrido ya no es necesario usarlo de nuevo por eso se deja en comentario


y_test_predict = lin_model.predict(X_test)
r2 =r2_score(y_test,y_test_predict)
print ("Thw model performance for testing set")
print ("-------------------------------------")
print ("R2 score in   {}".format(r2))   #r2 es el coheficiente de determinación o determinación múltiple que si está dentro delos parametros del .75 hasta antes de 1 está muy bien
```

```python
print ("Coefficients: \n", lin_model.coef_)
print (" Independent term: \n", lin_model.intercept_)
```

```python
# Recta de regresión lineal simple
experiencia =experiencia.values.reshape(-1,1)  # una vez que lo hemos corrido ya no es necesario usarlo de nuevo por eso se deja en comentario



predicted_data_salario = lin_model.predict(experiencia)
predicted_data_salario[0:5]
```

```python
plt.scatter(experiencia,salario)
plt.plot(experiencia,predicted_data_salario, color = "red")
plt.title ("Experiencia Vs Salario")
plt.xlabel ("Experiencia")
plt.ylabel("Salario")
plt.show()
```
