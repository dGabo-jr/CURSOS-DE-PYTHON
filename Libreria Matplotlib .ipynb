```python
#primer ejemplo

edad_x = [20,25,30,35,40,45,50,55,60,65]
salario_cd_y = [20000,30000,33000,40000,42000,45000,47000,60000,70000,80000]
salario_fed_y = [10000,15000,17000,27000,30000,32000,35000,40000,49000,50000]
salario_bed_y = [15000,20000,25000,27000,35000,40000,45000,50000,55000,60000]
```

```python
#nuevo grafico con datos existentes
import numpy as np

plt.plot(edad_x, salario_cd_y, label = "cientifico de datos", ) # Linestyle es estilo de linea  -- 
plt.plot(edad_x, salario_fed_y, label = " Front End desarrollador") 
plt.plot(edad_x, salario_bed_y, label = " Back End desarrollador")  # comparar  varios datos



salario_medio=np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis =0)/3
plt.plot (edad_x, salario_medio, label = "Salario medio",linestyle="--", color ="#000000")


plt.title ("Salario MEDIO en $USD  por Edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio %USD")
plt.grid(True)  # para poner la grafica cuadriculada

plt.legend() # para mostrar la leyenda por barra de grafico
plt.show()
```

```python


plt.plot(edad_x, salario_bed_y, label = " Back End desarrollador")  # comparar  varios datos



salario_medio_y=np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis =0)/3
plt.plot (edad_x, salario_medio, label = "Salario medio",linestyle="--", color ="#000000")

plt.fill_between(edad_x, salario_bed_y, salario_medio_y,
                 interpolate= True, alpha=0.33, label = "Between" )  #Rellenar huecos por enciman y por debajo de la media

plt.title ("Salario MEDIO en $USD  por Edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio %USD")
plt.grid(True)  # para poner la grafica cuadriculada

plt.legend() # para mostrar la leyenda por barra de grafico
plt.show()
```

```python
plt.plot(edad_x, salario_bed_y, label = " Back End desarrollador")  # comparar  varios datos



salario_medio_y=np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis =0)/3
plt.plot (edad_x, salario_medio, label = "Salario medio",linestyle="--", color ="#000000")

plt.fill_between(edad_x, salario_bed_y, salario_medio_y, where = (salario_bed_y > salario_medio_y),  # 
                 interpolate= True, alpha=0.33, label = "Between" )  #Rellenar huecos por encima de la media

plt.fill_between(edad_x, salario_bed_y, salario_medio_y, where = (salario_bed_y <= salario_medio_y),   
                 interpolate= True, color= "yellow", alpha=0.33, label = "Between" )  #Rellenar huecos  por debajo de la media


plt.title ("Salario MEDIO en $USD  por Edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio %USD")
plt.grid(True)  # para poner la grafica cuadriculada

plt.legend() # para mostrar la leyenda por barra de grafico
plt.show()
```
