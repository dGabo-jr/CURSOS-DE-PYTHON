```python
import matplotlib.pyplot as plt
import numpy as np

# Datos para los grficos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x)

x1 = [1, 2, 3, 4, 5]
y11= [2, 4, 1, 3, 5]

# Creacin de la figura y los subplots
fig, axs = plt.subplots(4, 4, figsize=(14, 14))

# -Grfico de lnea simple
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Grfico de lnea simple')

# Grfico de lnea con dos curvas
axs[0, 1].plot(x, y1, label='sin(x)')
axs[0, 1].plot(x, y2, label='cos(x)')
axs[0, 1].legend()
axs[0, 1].set_title('Grfico de lnea con dos curvas')

# 3.-Grfico de dispersin
axs[0, 2].scatter(x, y3)
axs[0, 2].set_title('Grfico de dispersin')

#  Grfico de barras vertical
axs[1, 0].bar(['A', 'B', 'C', 'D', 'E'], [3, 7, 1, 6, 4])
axs[1, 0].set_title('Grfico de barras vertical')

# Grfico de barras horizontal
axs[1, 1].barh(['A', 'B', 'C', 'D', 'E'], [3, 7, 1, 6, 4])
axs[1, 1].set_title('Grfico de barras horizontal')

# Grfico de pastel
axs[1, 2].pie([15, 30, 45, 10], labels=['A', 'B', 'C', 'D'])
axs[1, 2].set_title('Grfico de pastel')

# Grfico de rea
axs[2, 0].fill_between(x, y1, y2, alpha=0.5)
axs[2, 0].set_title('Grfico de rea')

# Grfico de contorno
X, Y = np.meshgrid(x, x)
Z = np.sin(X) * np.cos(Y)
axs[2, 1].contour(X, Y, Z)
axs[2, 1].set_title('Grfico de contorno')

# Grfico de mapa de calor
axs[2, 2].imshow(Z, cmap='hot', extent=[0, 10, 0, 10])
axs[2, 2].set_title('Grfico de mapa de calor')



# Crear grfico Stem
plt.stem(x1, y11)

# Configurar ejes
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Grfico Stem')

# Mostrar grfico
plt.show()
# Ajuste de la distribucin de los subplots
plt.tight_layout()

# Mostrar la figura
plt.show()
```

# **APLICACIONES PRINCIPALES DE CADA UNO DE LOS GRFICOS BSICOS**
**Grfico de lnea simple**: se utiliza para visualizar la evolucin de una variable a lo largo del tiempo
  Para mostrar la relacin entre dos variables. 

  EJEMPLO:  Representar la evolucin de las temperaturas a lo largo del ao en Salvatierra 

  image.png

**Grfico de lnea con dos curvas**: Se utiliza para comparar dos variables que evolucionan en funcin del tiempo. 
 
 Ejemplo:  Puede utilizarse para representar los indices de alumnos aprobados y reprobados en un Ciclo escolar o una carrera.
 image.png

**Grfico de dispersin:** se utiliza para visualizar la relacin entre dos variables continuas. 

EJEMPLO:  puede representar la relacin entre el precio de un producto y su demanda.
image.png

**Grfico de barras vertical:** se utiliza para comparar valores numricos discretos entre diferentes categoras. 
Por ejemplo, puede utilizarse para representar las diferentes calificaciones de un grupo en una materia.

image.png

**Grfico de barras horizontal:** se utiliza para comparar valores numricos discretos entre diferentes categoras, pero con los nombres de las categoras en el eje vertical en lugar del eje horizontal. Este tipo de grfico se utiliza a menudo para representar datos categricos.

Ejemplo: Calificaciones avances en la curricula (materias ) de una carrera universitaria

image.png

**Grfico de pastel:** se utiliza para mostrar la proporcin de diferentes categoras en una variable. 


Por ejemplo, puede utilizarse para representar la proporcin de diferentes tipos de usuarios en un sitio web.

**Grfico de rea:** se utiliza para visualizar la evolucin de una variable en el tiempo, mostrando el rea debajo de la curva. 


Por ejemplo, puede utilizarse para representar la evolucin de la superficie cubierta por un glaciar a lo largo del tiempo.

**Grfico de contorno:** se utiliza para visualizar los valores de una funcin en un plano 2D, utilizando curvas de nivel para representar los puntos de igual valor. 

Por ejemplo, puede utilizarse para representar la topografa de una regin.

image.png


**Grfico de mapa de calor:** se utiliza para representar valores numricos en un plano 2D, utilizando colores para mostrar la intensidad de los valores. 

Por ejemplo, puede utilizarse para representar la concentracin de un contaminante en diferentes zonas de una ciudad.

image.png

**GRFICO STEM:**  El grfico Stem tambin se utiliza para visualizar datos de seales, como seales de audio o de radiofrecuencia. En este caso, el grfico muestra la amplitud de la seal en el eje y y el tiempo o frecuencia en el eje x.

image.png

```python
import matplotlib.pyplot as plt

# Definimos los datos de las variables
n_variables = 8
porcentajes = [25, 30]

for i in range(n_variables-2):
    porcentajes.append(porcentajes[-1] + 5)

# Creamos la figura y los ejes
fig, ax = plt.subplots()

# Creamos el grfico de barras horizontales
ax.barh(range(n_variables), porcentajes, height=0.7, color='#007acc')

# Aadimos etiquetas a los ejes
ax.set_xlabel('Porcentaje')
ax.set_ylabel('Variable')

# Aadimos un ttulo al grfico
ax.set_title('Grfico de barras horizontales')

# Aadimos una lnea vertical para indicar el 50%
ax.axvline(x=50, color='gray', linestyle='--')

# Mostramos el grfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Creamos unos datos de ejemplo
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2 - Y**2)

# Creamos la figura y los ejes
fig, ax = plt.subplots()

# Creamos el grfico de contorno
cs = ax.contour(X, Y, Z)

# Aadimos etiquetas y ttulo
ax.clabel(cs, inline=1, fontsize=10)
ax.set_title('Grfico de contorno')

# Mostramos el grfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Creamos los datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# Creamos la figura y los ejes
fig, ax = plt.subplots()

# Creamos el grfico de mapa de calor
heatmap = ax.pcolormesh(X, Y, Z, cmap='RdBu')

# Aadimos una barra de color
fig.colorbar(heatmap)

# Aadimos etiquetas y ttulo
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Grfico de mapa de calor')

# Mostramos el grfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos del archivo CSV
empresas = pd.read_csv('datos_empresas.csv')

# Seleccionar solo las columnas de coordenadas y categora de cluster
coordenadas = empresas[['Latitud', 'Longitud', 'Cluster']]

# Agrupar los datos por categora de cluster y contar el nmero de empresas en cada categora
grupos_clusters = coordenadas.groupby(['Cluster']).count()

# Transformar los datos en una matriz para generar el mapa de calor
matriz_clusters = grupos_clusters.pivot_table(index='Longitud', columns='Latitud', values='Cluster')
# Generar el mapa de calor
plt.imshow(matriz_clusters, cmap='YlGnBu')

# Aadir etiquetas a los ejes
plt.xticks(range(len(matriz_clusters.columns)), matriz_clusters.columns)
plt.yticks(range(len(matriz_clusters.index)), matriz_clusters.index)

# Aadir un ttulo al grfico
plt.title('Mapa de calor de clusters industriales en Guanajuato')

# Mostrar el grfico
plt.show()

```
